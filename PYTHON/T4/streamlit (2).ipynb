```python
%%writefile pb711.py
import streamlit as st
import numpy as np
st.title("Streamlit profile app")
st.header("PB 711 PROG.")

name=st.text_input("enter name")
age=st.slider("select age",10,100)
gender=st.radio("choose gender",["male","female","other"])
hobby=st.multiselect("select hobby",["read","play","traveling "])
photo=st.file_uploader("upload pic ",type=["jpg","png","jpeg"])

if st.button("submit profile"):
    st.subheader("profile detail")
    st.write("name:",name)
    st.write("age:",age)
    st.write("gender:",gender)
    st.write("hobby:",", ".join(hobby))
    if photo:
        st.image(photo)
st.balloons()
```

```python
%%writefile pb713.py
import streamlit as st
import numpy as np
st.title("Streamlit profile app")
st.header("PB 713 PROG.")
c1,c2,c3,c4,c5=st.columns(5)

m1=c1.number_input("PYTHON",0,100,0)
m2=c2.number_input("FSD",0,100,0)
m3=c3.number_input("DE",0,100,0)
m4=c4.number_input("ETC",0,100,0)
m5=c5.number_input("PS",0,100,0)

if st.button("calculate mark"):
    total=m1+m2+m3+m4+m5
    avg=total/5
    
    if avg>=60:
        div="first"
    elif avg>=40:
        div="second"
    else:
        div="fail"
    with st.expander("result"):
        st.write("total",total)
        st.write("avg",avg)
        st.write("div",div)
```

```python
%%writefile pb714.py
import streamlit as st
import numpy as np
st.title("Streamlit profile app")
st.title("⚖️ BMI Calculator")
st.header("PB 714 PROG.")

st.write("Calculate your Body Mass Index (BMI) using weight in kilograms and height in centimeters.")

weight_kg = st.number_input("Enter your weight (kg):", min_value=1.0, step=0.1, format="%.2f")
height_cm = st.number_input("Enter your height (cm):", min_value=30.0, step=0.1, format="%.2f")

# --- Calculation ---
if st.button("Calculate BMI"):
    if height_cm > 0:
        # Convert height from cm to meters
        height_m = height_cm / 100
        # Calculate BMI: weight (kg) / height (m)^2
        bmi = weight_kg / (height_m ** 2)
        
        # --- Display Results ---
        st.subheader("Your BMI Result:")
        st.success(f"Your BMI is **{bmi:.2f}**") # Display BMI with 2 decimal places

        # --- BMI Interpretation ---
        if bmi < 18.5:
            st.warning("Category: Underweight")
        elif 18.5 <= bmi < 25:
            st.info("Category: Normal weight")
        elif 25 <= bmi < 30:
            st.info("Category: Overweight")
        else:
            st.error("Category: Obesity")
    else:
        st.error("Height must be greater than zero to calculate BMI.")

```

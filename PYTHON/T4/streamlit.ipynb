```python
%%writefile pb711.py
import streamlit as st
import numpy as np
st.title("Streamlit profile app")
st.header("PB 711 PROG.")

name=st.text_input("enter name")
age=st.slider("select age",10,100)
gender=st.radio("choose gender",["male","female","other"])
hobby=st.multiselect("select hobby",["read","play","traveling "])
photo=st.file_uploader("upload pic ",type=["jpg","png","jpeg"])

if st.button("submit profile"):
    st.subheader("profile detail")
    st.write("name:",name)
    st.write("age:",age)
    st.write("gender:",gender)
    st.write("hobby:",", ".join(hobby))
    if photo:
        st.image(photo)
st.balloons()
```

```python
%%writefile pb712.py
import streamlit as st
import numpy as np
st.title("Streamlit profile app")
st.header("PB 712 PROG.")
with st.sidebar:
    country=st.selectbox("select country",['india','usa','uk','canada'])
totalpopulation = st.number_input("total population")
totalvac = st.number_input("total vacinated")

if st.button("calculate %"):
    per=(totalvac/totalpopulation)*100
    st.progress(min(int(per),100))
    
    if per>=70:
        st.success("good")
    else:
        st.warning("bad")
```

```python
def find_most_prolific_sender(filename):
    counts = {}
    with open(filename) as file:
        for line in file:
            if line.startswith('From '):
                words = line.split()
                if len(words) > 1:
                    email = words[1]
                    counts[email] = counts.get(email, 0) + 1

    # Print the dictionary
    print(counts)

    # Find the sender with the maximum count
    max_sender = None
    max_count = None
    for sender, count in counts.items():
        if max_count is None or count > max_count:
            max_sender = sender
            max_count = count

    print(max_sender, max_count)


if __name__ == '__main__':
    find_most_prolific_sender('mbox-short.txt')
```

## Write a “pager” program. Your solution should prompt for a filename, and display the text file 25 lines at a time, pausing
each time to ask the user to enter the word “continue”, in order to show the next 25 lines or enter the word “stop” to
close the file.


```python
file=open("mbox-short.txt")
temp=True
line=file.readlines()
total=len(line)
start=0
end=min(25,total)
```

```python
while start<total:
    for i in range(start,end):
        print(line[i],end="")
        
        user=input("enter stop ")
        if user.lower()=="stop":
            break
        start=+25
        end=min(start+25,total)        
```
